```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import RobustScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor, AdaBoostRegressor
from sklearn.linear_model import LogisticRegression
```

```python
df_default = pd.read_csv('data/credit_card_default.csv')
df_default.head()
```

```python
#EDA

#make names in lower case
df_default.columns = df_default.columns.str.lower()
# For pay features if the <= 0 then it means it was not delayed. Set all negativ values to 0
pay_features = ['pay_0','pay_2','pay_3','pay_4','pay_5','pay_6']
for p in pay_features:
    df_default.loc[df_default[p]<=0, p] = 0
# Rename target column to just "default"
df_default.rename(columns={'default payment next month':'default'}, inplace=True)

df_default.head()
```

```python
# Create dummy variables
# Base values: female, other_education, not_married
# == return True or False and astype('int') coverts to 1 for True and 0 for False
df_default['grad_school'] = (df_default['education'] == 1).astype('int')
df_default['university'] = (df_default['education'] == 2).astype('int')
df_default['high_school'] = (df_default['education'] == 3).astype('int')
df_default['male'] = (df_default['sex']==1).astype('int')
df_default['married'] = (df_default['marriage'] == 1).astype('int')
# Remove columns
df_default.drop(['education', 'sex', 'marriage'] , axis=1, inplace=True)
```

```python
df_default.head()
```

```python
#Create train and test and scaling

target_name = 'default'
robust_scaler = RobustScaler()
x = df_default.drop('default', axis=1)
feature_names = x.columns
x = robust_scaler.fit_transform(x)
y = df_default[target_name]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=55)
```

```python
#Create df to save result

models = pd.DataFrame(index=['train_rmse', 'test_rmse'], 
                      columns=['LR', 'Bagging', 'RandomForest', 'Boosting'])
```
